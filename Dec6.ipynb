```python
#!/usr/bin/env python

from collections import Counter

def most_common_character(input_str):
    return ''.join([
            Counter(chars_in_pos).most_common(1)[0][0]
            for chars_in_pos in
            zip(*input_str.splitlines()[::-1])])

def least_common_character(input_str):
    return ''.join([
            Counter(chars_in_pos).most_common()[-1][0]
            for chars_in_pos in
            zip(*input_str.splitlines()[::-1])])
```
